```python
import numpy as np
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
X = loadmat("ex7/ex7data2.mat")['X']
df = pd.DataFrame(X)
df.columns=['X1', 'X2']
```

```python
sns.set()
```

```python
g = sns.lmplot(x='X1', y='X2', data=df, fit_reg=False)
g.fig.set_size_inches(8,6)
plt.show()
```

```python
def init_centroids(X, K):
    n, m = X.shape
    random_idx = np.random.randint(n, size=K)
    return X[random_idx, :].copy()
```

```python
def get_closest_idx(X, centroids):
    d = np.zeros(shape=(X.shape[0], K))
    for i in range(K):
        d[:,i] = np.sum((X - centroids[i,:])**2, axis=1)
    return d.argpartition(1)[:,0]
```

```python
def compute_centroids(X, idx, K):
    n = X.shape[1]
    centroids = np.zeros(shape=(K, n));
    for i in range(K):
        centroids[i,:] = np.mean(X[idx == i,:], axis=0)
    return centroids
```

```python
K = 3

centroids = init_centroids(X, K)
centroids

for i in range(1000):
    idx = get_closest_idx(X, centroids)
    centroids = compute_centroids(X, idx, K)
    
df['clusters'] = idx
g = sns.lmplot(x='X1', y='X2', data=df,hue='clusters', fit_reg=False)
g.fig.set_size_inches(8,6)
plt.title('Cluster after Iter {}'.format(i))
plt.show()
```
